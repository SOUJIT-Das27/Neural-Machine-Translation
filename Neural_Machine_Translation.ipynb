# Important Modules

```python
#importing Necessary modules

import pandas as pd

#importing deep learning modules

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM , Dense , Embedding
from sklearn.model_selection import train_test_split
```

# Dataset

Dataset consist of more than 10,000 collected data to make better model.Stored into .csv file

```python
#Loading Dataset

dataset = pd.read_csv("dataset.csv");
```

```python
#Glimpse of dataset

dataset.tail()
```

# Differenciating English And Bengali Words

 It's an approach of word preprocessing step of Natural Language Processing. 
         Here , tried to make speparate list for all english and bengali sentences for furthur process

```python

#differenciation of english and bengali sentences

english_sentences = dataset["Eng"].tolist()

bengali_sentences = dataset["Beng"].tolist()

print(english_sentences)
```

```python
#English word Vectorization

token_vector_english = Tokenizer()

token_vector_english.fit_on_texts(english_sentences)

english_vec = token_vector_english.texts_to_sequences(english_sentences)


#Sample Vectorized List

print(english_vec)
```

# Tokenization   And  Vectorization

```python
#Bengali word Vectorization

token_vector_bengali = Tokenizer()

token_vector_bengali.fit_on_texts(bengali_sentences)

bengali_vec = token_vector_bengali.texts_to_sequences(bengali_sentences)

#Sample Vectorized List

print(bengali_vec)
```

```python
#Extracting Vocabulary

vocabulary_size_english = len(token_vector_english.word_index) + 1

vocabulary_size_bengali = len(token_vector_bengali.word_index) + 1

print(vocabulary_size_english)
```

# Padding

```python
#it's a python shortcut for taking maximum sentence

max_length = max(len(sentence) for sentence in (english_vec + bengali_vec))

print(max_length)
```

```python
padded_english_vec = pad_sequences(english_vec, maxlen = max_length, padding = "post")

padded_bengali_vec = pad_sequences(bengali_vec, maxlen = max_length, padding = "post")

print(padded_bengali_vec)
```

# Making Neural Network

```python
embedding_dim = 256
units = 512
```

```python
#Encoder
encoder_inputs = Input(shape=(max_length,))
encoder_emb = Embedding(input_dim=vocabulary_size_english,output_dim = embedding_dim)(encoder_inputs)
encoder_LSTM = LSTM(units , return_state = True)
encoder_outputs,state_h,state_c = encoder_LSTM(encoder_emb)
encoder_states = [state_h,state_c]
```

```python
# Decoder

decoder_inputs = Input(shape=(max_length,))
dec_emb_layer = Embedding(input_dim=vocabulary_size_bengali, output_dim=embedding_dim)
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = LSTM(units, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)
decoder_dense = Dense(vocabulary_size_bengali, activation='softmax')
output = decoder_dense(decoder_outputs)
```

# Deep Learning Model Creation

```python
# Model
model = Model([encoder_inputs, decoder_inputs], output)
```

```python
# Compilation of model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
X_train, X_val, y_train, y_val = train_test_split(padded_english_vec, padded_bengali_vec, test_size=0.2)
```

```python
model.fit([X_train, X_train], y_train, validation_data=([X_val, X_val], y_val), epochs=100, batch_size=64)
```
